```python
archivo.groupby(['Canton'])['Nombre'].count()
```

```python
num_colegios = archivo.groupby(['Canton'])['Nombre'].count().tolist()
```

```python
num_colegios
```

```python
cantones={'CALVAS':num_colegios[0],
          'CATAMAYO':num_colegios[1],
          'CELICA':num_colegios[2],
          'CHAGUARPAMBA':num_colegios[3],
          'ESPINDOLA':num_colegios[4],
          'GONZANAMA':num_colegios[5],
          'LOJA':num_colegios[6],
          'MACARA':num_colegios[7],
          'OLMEDO':num_colegios[8],
          'PALTAS':num_colegios[9],
          'PINDAL':num_colegios[10],
          'PUYANGO':num_colegios[11],
          'QUILANGA':num_colegios[12],
          'SARAGURO':num_colegios[13],
          'SOZORANGA':num_colegios[14],
          'ZAPOTILLO':num_colegios[15]
         }

data = pd.Series(cantones).reset_index(name='value').rename(columns={'index':'cantones'})
data['angle'] = data['value']/data['value'].sum() * 2*pi
data['color'] = Category20c[len(cantones)]

p = figure(plot_height=350, title="Porcentaje de colegios de cada cantón de la provincia de Loja", toolbar_location=None,
           tools="hover", tooltips="@cantones: @value", x_range=(-0.5, 1.0))

p.wedge(x=0, y=1, radius=0.4,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', legend='cantones', source=data)

p.axis.axis_label=None
p.axis.visible=False
p.grid.grid_line_color = None

show(p)
```

```python
archivo.groupby(['Regimen Escolar'])['Nombre'].count()
```

```python
num_regimen = archivo.groupby(['Regimen Escolar'])['Nombre'].count().tolist()
num_regimen
```

```python
output_file("colegios_por_regimen.html")

regimen = ['Sierra', 'Costa']
counts = [num_regimen[0], num_regimen[1]]

source = ColumnDataSource(data=dict(regimen=regimen, counts=counts, color=Spectral6))

p = figure(x_range=regimen, y_range=(0,100), plot_height=700, title="Número de colegios pertenecientes a la región Costa y Sierra de la provincia de Loja, 2015",
           toolbar_location=None, tools="")

p.vbar(x='regimen', top='counts', width=0.5, color='color', legend="regimen", source=source)

p.xgrid.grid_line_color = None
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

show(p)
```
